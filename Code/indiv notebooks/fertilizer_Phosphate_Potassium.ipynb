```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
phosphate_path = "../data/phosphate fertilizer use per hectare of cropland 2017.csv"
potassium_path = "../data/potassium use per hectare of cropland 2017.csv"



phosphate_df = pd.read_csv(phosphate_path)
potassium_df = pd.read_csv(potassium_path)


combined_df = pd.merge(phosphate_df, potassium_df, how='outer')

combined_df
```

```python
combined_df = combined_df.rename(columns=
                                 {'Entity':'Country/Continent',
                                  'Fertilizers indicators - Nutrient phosphate P2O5 (total) - 3103 - Use per area of cropland - 5159 - kg/ha':'Phosphate_use_per_area_kg_per_ha',
                                  'Fertilizers indicators - Nutrient potash K2O (total) - 3104 - Use per area of cropland - 5159 - kg/ha':'Potah_use_per_area_kg_per_ha'})
combined_df
```

```python
combined_df = combined_df.loc[combined_df['Phosphate_use_per_area_kg_per_ha'] > 0.1, ]
combined_df
```

```python
combined_filterd_df= combined_df[combined_df['Potah_use_per_area_kg_per_ha'] < 0.1]

combined_filterd_df
```

```python
combined_filterd_df.to_csv('combined_fertilizer')
```
